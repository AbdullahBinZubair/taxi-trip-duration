```python
df = pd.read_csv('../input/train.csv')
test = pd.read_csv('../input/test.csv')
Y = df.pop('trip_duration')
df.pop('id')
df.pop('pickup_datetime')
# test.pop('id')
# test.pop('pickup_datetime')
print ("Hello")
```

```python
df['store_and_fwd_flag'][df['store_and_fwd_flag']=='N'] = 0
df['store_and_fwd_flag'][df['store_and_fwd_flag']=='Y'] = 1
test['store_and_fwd_flag'][test['store_and_fwd_flag']=='N'] = 0
test['store_and_fwd_flag'][test['store_and_fwd_flag']=='Y'] = 1
```

```python
X = np.array(df)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
clf = MLPRegressor(alpha=0.001, hidden_layer_sizes = (10,), max_iter = 50000, 
                 activation = 'logistic', verbose = 'True', learning_rate = 'adaptive')
a = clf.fit(X_train, y_train)
```

```python
pred = clf.predict(X_test)
print (np.sqrt(np.sum((np.round(pred) - y_test)**2)))
```

```python
res = clf.predict(np.array(test))
res = np.round(res)
```

```python
array = np.vstack((np.array(test['id']),res))
array = array.T
pd.DataFrame(array).to_csv("submit.csv")
# print (array)
```

```python
from IPython.display import HTML
import pandas as pd
import numpy as np
import base64

# function that takes in a dataframe and creates a text link to  
# download it (will only work for files < 2MB or so)
def create_download_link(df, title = "Download CSV file", filename = "data.csv"):  
    csv = df.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
```

```python
create_download_link(pd.DataFrame(array))
```
